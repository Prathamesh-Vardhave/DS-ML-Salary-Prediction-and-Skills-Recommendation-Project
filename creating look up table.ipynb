```python
import pandas as pd
import joblib
```

```python
df=pd.read_csv("updated_final_dataset.csv")
```

```python
job_role_map = df.groupby("Job_Role")["job_role_encoded"].mean().to_dict()
joblib.dump(job_role_map, "job_role_map.pkl")


city_salary_map = df.groupby("primary_city")["city_salary_index"].mean().to_dict()
joblib.dump(city_salary_map, "city_salary_map.pkl")


city_group_map = df.groupby("primary_city")["city_group"].first().to_dict()
joblib.dump(city_group_map, "city_group_map.pkl")
```

```python
df_rec = df[['Job_Role', 'exp_bucket', 'skills']].copy()
df_rec = df_rec[df_rec['skills'].notna()]

joblib.dump(df_rec, "recommendation_data.pkl")
```
